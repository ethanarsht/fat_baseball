```python
import pandas as pd
import numpy as np
from bokeh.io import output_file, show, save
from bokeh.layouts import widgetbox, gridplot
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.models.widgets import TableColumn, DataTable
from bokeh.plotting import figure
```

```python
df_bokeh = pd.read_csv('bokeh_chart.csv', index_col = 0)
df_bokeh
```

```python
df_years = df_bokeh.groupby('name').count().iloc[:, :1].reset_index()
```

```python
df_chart = pd.merge(df_bokeh, df_years, on = 'name', how = 'outer').rename(columns = {'year_y':'yoc'})
```

```python
df_chart = df_chart[df_chart['year_x'] == 2018]
```

```python
df_chart = df_chart[~df_chart.name.isin(['Scooter Gennett','Chris Davis','Leonys Martin', 'Pablo Sandoval'])]
df_chart['1'] = df_chart['1'].round(2)
```

```python
df_chart.head()
```

```python
df_chart[df_chart.name.str.contains('Gennett')]
```

```python
output_file('docs/index.html', title = 'Last-year index dashboard')
source = ColumnDataSource(df_chart)
columns = [TableColumn(field = 'name', title = 'Player'),
           TableColumn(field = '1', title = 'Last Year Index')]
data_table = DataTable(source = source, columns = columns)




hover = HoverTool(tooltips = (
[
    ('Name', '@name'),
    ('BMI', '@bmi'),
    ('Years since debut', '@yoc'),]))

tools = [hover, 'box_zoom', 'tap']
p = figure(
    title = 'BMI vs Career Longevity for Active Pitchers',
    plot_width = 800,
    plot_height = 600,
    tools = tools
)
p.yaxis.axis_label = 'Years since MLB debut'
p.xaxis.axis_label = 'BMI'


           
p.scatter(x = 'bmi', y = 'yoc', source = source, size = 8)
save(p)

grid = gridplot([[p, widgetbox(data_table)]])
save(grid)

```
